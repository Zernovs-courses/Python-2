```python
df.loc[df['Age'] < 18]['AgeGroup'] = 'Child'
```

```python
def create_age_group(age):
    if age < 18:
        return 'Child'
    else:
        return 'Adult'
    
df['AgeGroup'] = df['Age'].apply(create_age_group)
```

```python
df
```

```python
df.groupby(['Pclass', 'Sex', 'AgeGroup'])['Survived'].mean()
```
