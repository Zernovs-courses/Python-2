```python
df['AgeGroup'] = ''
df.head()
```

```python
def create_age_group(age):
    if age < 18:
        return 'child'
    else:
        return 'adult'
    
df['AgeGroup'] = df['Age'].apply(create_age_group)
df.head()
```
